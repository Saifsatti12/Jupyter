```python
import numpy as hs
```

```python
import matplotlib.pyplot as plt
```

```python
x = hs.linspace(-5,4,10)
y = x**2
```

```python
x
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('velocity',fontsize=20)
plt.ylabel('acceleration',fontsize=18)
```
